```python
from pathlib import Path
from encoder import raw_data_ready, get_raw_txt_file_paths

assert raw_data_ready()
raw_txt_paths: dict[int, Path] = get_raw_txt_file_paths()
```

```python
raw_contents: dict[int, str] = {}

for year, txt_path in raw_txt_paths.items():
  with txt_path.open('r') as f:
    raw_contents[year] = f.read()
  print(f'{year}: {len(raw_contents[year])} bytes')
```

```python
daylight_saving_count = 0
all_count = 0
for year, content in raw_contents.items():
  if '夏令時間' in content:
    daylight_saving_count += 1
    print(f'夏令時間 is on for year {year}')
  all_count += 1

print(f'In the total of {all_count} years, daylight saving time is used in {daylight_saving_count} years.')
```

```python
def is_valid_line(line: str) -> bool:
  if '年' not in line:
    return False
  if '月' not in line:
    return False
  if '日' not in line:
    return False
  if '夏令時間' in line:
    return False
  return True

for year, content in raw_contents.items():
  lines = content.splitlines()
  valid_lines = [l for l in lines if is_valid_line(l)]
  assert len(valid_lines) in [365, 366], f'Found {len(valid_lines)} days in {year}'

  jieqi_count = 0
  for line in valid_lines:
    splitted = line.split()
    assert len(splitted) in [3, 4]
    if len(splitted) == 4:
      jieqi_count += 1
  
  assert jieqi_count == 24
```
